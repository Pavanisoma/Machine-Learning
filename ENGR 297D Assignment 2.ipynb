```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False 


true_w = np.array([3])
inp = []
output = []
d = len(true_w)
points = []

for i in range(10000):
    x = np.random.randn(d)
    y = true_w.dot(x)+np.random.randn()
    points.append((x,y))
    inp.append(x)
    output.append(y)

inp = np.asarray(inp)
plt.scatter(inp[:, 0], output, s=0.1, color='k', marker='o')
plt.plot([-3.3,3.3],[-10,10])
```

```python
def F(w):
    value = 0
    for x,y in points:
        value += (x.dot(w) - y) ** 2
    
    
    return value/len(points)
```

```python

def dF(w):
    value = 0
    for x,y in points:
        value += 2*(x.dot(w)-y)*x
    return value/len(points)
```

```python
def gradientDescent(F,dF,d):
    w = np.zeros(d)
    l = 0.01 #step size
    for t in range(500):
        value = F(w)
        grad = dF(w)
        w = w - l*grad #Taking one step
        print('iteration {}, w = {}, F(q) = {}, F\'(w) = {}'.format(t,w,value,grad))
```

```python
gradientDescent(F,dF,d)
```
