```python
import pandas as pd 
import numpy as np
```

```python
def load_data():
    data = pd.read_csv("data.csv",header=None) 
    return data
```

```python
data = load_data()
```

```python
points=np.asarray(data)
```

```python
points
```

```python
def sF(w,i):
    x,y =points[i]
    value=(x*(w)-y)**2
    return value
```

```python
def sdF(w,i):
    x,y =points[i]
    value= 2*(x*(w)-y)*x
    return value
```

```python
def stochasticgradientD(sF,sdF):
    w=np.zeros(1)
    l=0.01
    for i in range(50):
        for j in range (len(points)):
            sf=sF(w,i)
            sg=sdF(w,i)
            w=w-l*sg
        print("iteration:{},w:{},f:{},g:{}".format(i,w,sf,sg))
```
